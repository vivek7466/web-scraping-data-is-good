```python
import requests
from bs4 import BeautifulSoup

# URL of the page to scrape
url = "https://quotes.toscrape.com/page/1/"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract information from the page
    quotes = soup.find_all('span', class_='text')
    authors = soup.find_all('small', class_='author')
    tags = soup.find_all('div', class_='tags')

    # Print the extracted information
    for i in range(len(quotes)):
        print("Quote:", quotes[i].get_text())
        print("Author:", authors[i].get_text())
        tags_list = [tag.get_text() for tag in tags[i].find_all('a', class_='tag')]
        print("Tags:", ', '.join(tags_list))
        print("------")

else:
    print("Failed to retrieve the page. Status code:", response.status_code)
```

```python
# URL of the page to scrape
url = "https://quotes.toscrape.com/page/1/"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract information from the page
    quotes = [quote.get_text() for quote in soup.find_all('span', class_='text')]
    authors = [author.get_text() for author in soup.find_all('small', class_='author')]
    tags = [', '.join(tag.get_text() for tag in div.find_all('a', class_='tag')) for div in soup.find_all('div', class_='tags')]

    # Create a DataFrame
    df = pd.DataFrame({'Quote': quotes, 'Author': authors, 'Tags': tags})

    # Print the DataFrame
    print(df)

else:
    print("Failed to retrieve the page. Status code:", response.status_code)
```
